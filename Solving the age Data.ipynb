#Computing Percentile

```python
arr = np.percentile(df.height,[25,50,75])
arr
```

# calculate IQR

```python
iqr = arr[2]-arr[0]
iqr
```

# Finding Outliers

```python
df.height[((df.height) < arr[0] - iqr) | (df.height) > arr[2]-iqr]
```

```python
sns.boxplot(df.height)
```

# Computing Standard Deviation and Variance

```python
print("variance is", round(np.var(df.height),2))
```

```python
print("Standard Deviation is", round(np.std(df.height),2))
```
